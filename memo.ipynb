# Filter PDFs that are Memorandums
Here, the first word is "Memorandum".

```python
from glob import glob
import pymupdf, os, shutil
```

```python
pdfs = glob("pdfs/*.pdf")
```

```python
if not os.path.exists("pdfs/memo"):
    os.makedirs("pdfs/memo")

def find_first_alphanumeric_word(text):
    # Split the text into words
    words = text.split()
    # Use a regular expression to find the first alphanumeric word
    for word in words:
        if word.isalnum():
            return word
    return None

def get_first_word_from_pdf(pdf_path):
    # Open the PDF file
    document = pymupdf.open(pdf_path)
    
    # Extract text from the first page
    first_page = document.load_page(0)
    text = first_page.get_text("text")
    
    # Find the first alphanumeric word
    first_word = find_first_alphanumeric_word(text)
    if first_word and first_word.lower() == "memorandum":
        # Move the file to the target folder
        shutil.copy(pdf_path, os.path.join("pdfs/memo", os.path.basename(pdf_path)))
        print(f"File '{pdf_path}' contains 'Memorandum' and has been copied to 'pdfs/memo'.")
    else:
        None

# Get the first word from the PDF
for x in pdfs:
    get_first_word_from_pdf(x)
```

```python
memos = glob("pdfs/memo/*")
```

```python
for x in memos:
    os.remove(os.path.join("pdfs", os.path.basename(x)))
```
