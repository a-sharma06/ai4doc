```python
import pytesseract, cv2, pymupdf
from pytesseract import Output
from PIL import Image
import numpy as np
```

```python
# Path to your PDF file
pdf_path = r"C:\Users\ashar\repos\ai4doc\pdfs\4606-3173.pdf"

# Open the PDF file
pdf_document = pymupdf.open(pdf_path)

# Iterate over each page
for page_num in [0]: #range(len(pdf_document)):
    # Select the page
    page = pdf_document.load_page(page_num)
    
    # Convert page to image
    pix = page.get_pixmap()

    pix.save(r"C:\Users\ashar\repos\ai4doc\other_scripts\output_image.png")
    
```

```python
# Path to the image file
image_path = r"C:\Users\ashar\repos\ai4doc\other_scripts\output_image.png"

# Load the image with PIL
image = Image.open(image_path)

# Convert the image to RGB (if it's not in that format)
image = image.convert('RGB')

# Use pytesseract to get bounding boxes
# 'output_type' set to 'dict' returns detailed information
boxes = pytesseract.image_to_boxes(image, output_type=pytesseract.Output.DICT)

# Load the image with OpenCV to visualize the bounding boxes
image_cv = cv2.imread(image_path)

# Get image dimensions
(h, w) = image_cv.shape[:2]

# Iterate through each box and draw rectangles
for i in range(len(boxes['char'])):
    # Get the bounding box coordinates
    x, y, w, h = int(boxes['left'][i]), int(boxes['bottom'][i]), int(boxes['right'][i]), int(boxes['top'][i])
    
    # Draw the bounding box on the image
    cv2.rectangle(image_cv, (x, h-y), (w, h-h), (0, 255, 0), 2)
```

```python
# Optionally save the result image
cv2.imwrite('image_with_boxes.png', image_cv)
```

```python
# Save or display the result image
cv2.imshow('Image with Bounding Boxes', image_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
